```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# データの読み込み
df_raw_gyro = pd.read_csv("data/05/gyroscope.csv")
df_raw_accel = pd.read_csv("data/05/accelerometer.csv")

# データの確認
print(df_raw_gyro.head())
print(df_raw_accel.head())

# ピークの検出
peaks, _ = find_peaks(df_raw_accel["y"], height=0)  # heightパラメータは適切に調整

# プロット設定
plt.subplots_adjust(hspace=0.3)
fig = plt.figure(figsize=(10, 8))

# 加速度センサデータのプロット（線と点）
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(df_raw_accel["t"], df_raw_accel["y"], label="加速度センサデータ")
ax1.plot(df_raw_accel["t"].iloc[peaks], df_raw_accel["y"].iloc[peaks], "x", label="極大値", markersize=4, color="red")
ax1.set_title("極大値に目印をつけた加速度センサデータ", fontsize=20)
ax1.set_xlabel("time[s]")
ax1.set_ylabel("加速度 Y[m/s²]")  # 単位を変更
ax1.legend()

plt.show()

# ジャイロセンサデータのプロット設定
plt.subplots_adjust(hspace=0.3)
fig = plt.figure(figsize=(10, 8))

ax2 = fig.add_subplot(2, 1, 2)
ax2.plot(df_raw_gyro["t"], df_raw_gyro["x"], label="ジャイロセンサデータ")
ax2.set_title("ジャイロセンサ", fontsize=20)
ax2.set_xlabel("time[s]")
ax2.set_ylabel("角速度 X[rad/s]")  # 単位を変更
ax2.legend()

plt.show()
```
