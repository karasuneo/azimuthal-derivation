```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# 
df_raw_gyro = pd.read_csv("data/05/gyroscope.csv")
df_raw_accel = pd.read_csv("data/05/accelerometer.csv")

# 
print(df_raw_gyro.head())
print(df_raw_accel.head())

# 
peaks, _ = find_peaks(df_raw_accel["y"], height=0)  # height

# 
plt.subplots_adjust(hspace=0.3)
fig = plt.figure(figsize=(10, 8))

# 
ax1 = fig.add_subplot(2, 1, 1)
ax1.plot(df_raw_accel["t"], df_raw_accel["y"], label="")
ax1.plot(df_raw_accel["t"].iloc[peaks], df_raw_accel["y"].iloc[peaks], "x", label="", markersize=4, color="red")
ax1.set_title("", fontsize=20)
ax1.set_xlabel("time[s]")
ax1.set_ylabel(" Y[m/s]")  # 
ax1.legend()

plt.show()

# 
plt.subplots_adjust(hspace=0.3)
fig = plt.figure(figsize=(10, 8))

ax2 = fig.add_subplot(2, 1, 2)
ax2.plot(df_raw_gyro["t"], df_raw_gyro["x"], label="")
ax2.set_title("", fontsize=20)
ax2.set_xlabel("time[s]")
ax2.set_ylabel(" X[rad/s]")  # 
ax2.legend()

plt.show()
```
