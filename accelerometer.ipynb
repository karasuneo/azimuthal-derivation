```python
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
```

```python
df = pd.read_csv("./data/07/accelerometer.csv")

# 
def plot_acceleration_data(start_time=None, end_time=None):
    # 
    if start_time is not None:
        df_filtered = df[df['t'] >= start_time]
    else:
        df_filtered = df
        
    if end_time is not None:
        df_filtered = df_filtered[df_filtered['t'] <= end_time]
    
    # X, Y, Z 
    plt.figure(figsize=(10, 6))
    plt.plot(df_filtered['t'], df_filtered['x'], label='X', marker='o')
    plt.plot(df_filtered['t'], df_filtered['y'], label='Y', marker='o')
    plt.plot(df_filtered['t'], df_filtered['z'], label='Z', marker='o')

    # 
    plt.title('Accelerometer Data Over Time')
    plt.xlabel('Time (s)')
    plt.ylabel('Acceleration (m/s^2)')
    plt.legend()
    plt.grid(True)
    plt.show()

# 
plot_acceleration_data(start_time=20.0, end_time=25.0)
```
